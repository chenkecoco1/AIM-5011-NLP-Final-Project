## Meme Text Processing

This process is based on all the json files generated from Google API. To facilitate the manipulation, all the json files were downloaded to local folders, regarding all labeled memes and hate memes, which is for labeling column information. Considering the text of the memes are not in a sentence structure, API extraction may cause bias, and hate memes regarding politics has country code and name infos, the text will be set in uppercase. 

```python
# import OS module
import os
 
# Get all file names
path = r"C:\Users\chenkecoco1\Documents\GitHub\AIM-NN\text-20211220T225138Z-001\text"
dir_list = os.listdir(path)

# check the first file name
print(dir_list[0])
```

```python
# get hate file names, which should be labeled as 1, hate
path_1 = r"C:\Users\chenkecoco1\Documents\GitHub\AIM-NN\Hate-20211222T004727Z-001\Hate"
dir_list_HATE = os.listdir(path_1)

# check the first file
print(dir_list_HATE[0])
```

```python
print(len(dir_list))
print(len(dir_list_HATE))
```

Files are not missing. After investigating the json file, text info were found in the "discription" part in each json file. The text will be extracted step by step from each json file.

```python
# set rules to reach the description data
import re
pattern = re.compile(r'description:+\D+[A-Z,a-z]') 
```

```python
# loop each file in the text folder and get all description info of each file

msg=[]

for name in dir_list:
    
    file = open(r'C:\Users\chenkecoco1\Documents\GitHub\AIM-NN\text-20211220T225138Z-001\text' + '\\' + name, 'r')
    Lines = file.readlines()
    s=' '
    for i in Lines:
        if bool(re.findall(pattern, i)):
            a = re.findall(pattern, i)
            s += ' '+str(a)
    msg.append(s)
```

No missing item were found. However, the text location of a meme are different. This cause a very irregular format of text. Therefore, the raw data caught from the json files need to be polished.

```python
# clean several round with regex
org_text=[]
for x in msg:
    # separate the text with "description:" as a mark
    first=' '.join(re.split(r'description:',x))
    # take off the '\\\\n'
    second=' '.join(re.split(r'\\\\n',first))
    # take off the non-character contents
    third=' '.join(re.split(r'[^a-zA-Z\d:]',second))
    # take off the unwanted spaces link the words
    fourth=' '.join(re.findall(r'\w+',third))
    # uppercase the text
    fifth = fourth.upper()
    
    org_text.append(fifth)
```

To match the "hate" meme file names, the .json extension needs to be taken off.

```python
# for matching the original memes names, the .json extension needs to be taken off
image_names=[]
for file in dir_list:
    image_names.append(file[:-5])
    
hate_names=[]
for h in dir_list_HATE:
    hate_names.append(h)
```

```python
image_names[0],hate_names[0]
```

File names are in the same format. The data needs to be stored in a dataframe. There are supposed to be 3 columns, file names, text and label.

```python
import pandas as pd

# file names as first column
df = pd.DataFrame(image_names,columns=['image_name'])
# text as second column
df['text']=pd.DataFrame(org_text)
```

We still need to label all the images labeled as hate (0 as not hate, 1 as hate). The hate item names are in the hate memes folder path. If the hate memes file name found, it will be labled as hate. 

```python
labels=[]

for name in image_names:
    if name in hate_names:
        labels.append(1)
    else:
        labels.append(0)
```

```python
# check first file labels
labels[0:5]
```

Labels will be added in the dataframe in a new column.

```python
df['label']=pd.DataFrame(labels)
```
